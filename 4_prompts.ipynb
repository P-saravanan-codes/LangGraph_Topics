#### **Prompts in LangGraph**

```python
from dotenv import load_dotenv
import os 
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
if os.environ["GOOGLE_API_KEY"]:
    print("API_key is set")
else:
    print("API_key is not set")
```

```python
llm = ChatGoogleGenerativeAI(model= "gemini-2.5-flash",temperature = 0)
llm
```

```python
my_prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant that helps to create and curate posts for a social media platform."),
        ("human","I want to create a post about{topic}")
    ]
)
```

```python
prompt_filled = my_prompt.invoke({"topic":"The importance of data privacy in the digital age"})
prompt_filled 
```

```python
llm.invoke(prompt_filled)
```

### Chain with prompts

```python
chain = my_prompt | llm
chain.invoke({"topic":"The importance of data privacy in the digital age"})
```
