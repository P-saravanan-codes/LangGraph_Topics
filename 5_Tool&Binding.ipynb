#### **Tools**

```python
from dotenv import load_dotenv
import os 
load_dotenv()

if os.environ["GOOGLE_API_KEY"]:
    print("API_key is set")
else:
    print("API_key is not set")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.tools import tool

llm = ChatGoogleGenerativeAI(model="gemini-2.0-pro", temperature=0.9)
```

```python
### TOOls 

from langchain_community.tools import DuckDuckGoSearchRun
@tool
def ddgs_tool(query:str) -> str:
    """This tool allows you to searches DuckDuckgo search for the given query and return the result."""
    duck_search = DuckDuckGoSearchRun()
    return duck_search.invoke("Trending news about langchain")
```

```python
from langchain_community.tools import ArxivQueryRun
from langchain_community.utilities import ArxivAPIWrapper

@tool
def arxiv_tool(query:str)->str:
    """This tool allows you to query the arxiv database for research papers."""
    arxiv_query = ArxivQueryRun(api_wrapper=ArxivAPIWrapper())
    return arxiv_query.invoke(query)

```

```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

@tool
def wiki_tool(query:str):
    """This tool allow you to search wikipedia for information on a given topic."""
    wiki_query = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
    return wiki_query.invoke("What is Langchain?")
```

```python
from langchain.tools import tool
@tool
def personal_info(name:str):
    """Use this tool to get personal information about a person.
    """

    info = {
        "Alice": "Alice is a software engineer with 5 years of experience in machine learning.",
        "Bob": "Bob is a data scientist who specializes in natural language processing.",
        "Charlie": "Charlie is a product manager with a background in AI and cloud computing."
    }
    return info.get(name, "No information available for this person.")
```

### Tool Binding

```python
tools = [ddgs_tool, arxiv_tool, wiki_tool, personal_info]
llm_with_binding = llm.bind_tools(tools)
```
